```python
files = !find data -name "bulletinb.*"
files = [Path(f) for f in files if not f.endswith('.pdf')]

result = []
for f in files:
    reader = ext.BulletinBReader(f)
    try:
        metadata = reader.extract_metadata()
    except Exception as err:
        warn(f"{f}: {str(err)}")
    result.append(
        {
            'name': f.name,
            'validity_start': metadata.validity_start,
            'validity_stop': metadata.validity_stop,
            'creation_date': metadata.creation_date
        }
    )
    
df = pd.DataFrame(result)
df.set_index('name', drop=True, inplace=True)
df
```
